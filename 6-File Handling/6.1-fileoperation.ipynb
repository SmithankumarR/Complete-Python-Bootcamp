#### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
### Read a Whole File

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## .strip() removes the newline character
```

```python
## Writing a file(Overwriting)

with open('example.txt','w') as file: # 'w' mode will overwrite the file if it already exists
    file.write('I started writing to the file.\n')
    file.write('this is a new line. i wrote just now.\n')
```

```python
## Write a file(without Overwriting)
with open('example.txt','a') as file: # 'a' mode will append to the file if it already exists, otherwise it will create a new file
    file.write("Append operation taking place!\n")
```

```python
### Writing a list of lines to a file
lines=['one more lines \n','today 2nd line \n','Third line time\n']
with open('example.txt','a') as file:
    file.writelines(lines) # writelines() method writes a list of lines to the file. It does not add newline characters, 
    # so we need to include them in the list of lines.
```

```python
### Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file: # 'wb'(write binary mode) opens the file for writing in binary mode
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content froma  source text fiile and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
#Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines) # line_count is the number of lines in the file, which is the length of the list of lines
        word_count = sum(len(line.split()) for line in lines) # word_count is the total number of words in the file, which is the sum of the number of words in each line.
        char_count = sum(len(line) for line in lines) # char_count is the total number of characters in the file, which is the sum of the number of characters in each line.
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

The w+ mode in Python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file exists, its content is truncated (i.e., the file is overwritten).

```python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the content of the file
    content=file.read()
    print(content)
```
