##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.



```python
## syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
## why functions?
num=332
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
## Call this function
even_or_odd(24)
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

result=add(2,4)
print(result)
    
```

```python
def miniCalculator(a,b,operator):
    if operator == "+":
        return a+b
    elif operator == "-":
        return a-b
    elif operator == "*":
        return a*b
    elif operator == "/":
        return a/b
    else:
        return "Invalid operator"   
result=miniCalculator(10,5,"*")
print(f"selected operator result: {result}")   
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome To the paradise")

greet("Charlie")
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome To the paradise")

greet()
```

```python
### Variable Length Arguments -- *args and **kwargs
## Positional And Keywords arguments

def print_numbers(*kan):
    for number in kan:
        print(number)
```

```python
print_numbers(1,4,5,8,"Charlie") # positional arguments --- more than one argument (*args)
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"smityh") # positional arguments --- more than one argument (*args)
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Charlie",age="23",country="India") # keyword arguments --- more than one key and value pairs  (**kwargs)
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Charlie",name="Charlie",age="23",country="India")
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
### Return multiple parameters
def multiply(a,b):
    return a*b,a-b,a+b,a/b,a,b

multiply(2,3)
```

```python
def palindrome(name):
    if name == name[::-1]:
        return f"{name} is a palindrome"
    else:  
        return f"{name} is not a palindrome"

palindrome("madam")
```

```python
def fabonacci(n):
    a=0
    b=1
    for i in range(n):
        print(a)
        a,b=b,a+b
fabonacci(5)
```

```python
def factorialnumber(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorialnumber(n-1)  
factorialnumber(5)
```
